```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
scores = pd.DataFrame({'scores':[9,6,7,6,6,9,7,8,8,8,6,8,6,6,10,8,6,7,7,8,8]})
sns.distplot(scores)
scores.describe()
```

```python
scores['z-score'] = (scores.scores-scores.scores.mean())/scores.scores.std()
scores['adjusted_score'] = np.round(1.5*scores['z-score'] + 12)
scores
```
